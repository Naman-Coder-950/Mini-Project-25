```python
import cv2
from google.colab.patches import cv2_imshow

# Input and output video paths
input_path = '/content/input_video.mp4'
output_path = 'output_video_gray.mp4'

# Open input video
cap = cv2.VideoCapture(input_path)

# Check if video opened successfully
if not cap.isOpened():
    print("❌ Error: Could not open input video.")
    exit()

# Get video properties
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Create VideoWriter for grayscale video
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height), isColor=False)

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert frame to grayscale
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Write frame to output
    out.write(gray_frame)

    # Display the FIRST FEW frames as preview
    if frame_count < 5:
        cv2_imshow(gray_frame)
    frame_count += 1

cap.release()
out.release()

print("✅ Conversion complete! Saved as:", output_path)
```
